We import libraries

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show,output_notebook
output_notebook()
```

It is given a <code>make_dashboard</code> function

```python
def make_dashboard(x, gdp_change, unemployment, title, file_name):
    output_file(file_name)
    p = figure(title=title, x_axis_label='year', y_axis_label='%')
    p.line(x.squeeze(), gdp_change.squeeze(), color="firebrick", line_width=4, legend="% GDP change")
    p.line(x.squeeze(), unemployment.squeeze(), line_width=4, legend="% unemployed")
    show(p)
```

The data is in a csv file at the following link

```python
links={'GDP':'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/clean_gdp.csv',\
       'unemployment':'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/clean_unemployment.csv'}
```

4. We create a dataframe of the column date from the dataframe of the GDP data

```python
x = dfGDP[['date']]
x.head()
```

We create a dataframe with the column change-current from the GDP data

```python
gdp_change = dfGDP[['change-current']]
gdp_change.head()
```

```python
unemployment = dfUnemp[['unemployment']]
```

```python
title = 'Dashboard'
```

```python
make_dashboard(x, gdp_change, unemployment, title, file_name='dashboard.html')
```
